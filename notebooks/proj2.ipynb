```python
df = df.drop_duplicates()
```

```python
df["Order Date"] = pd.to_datetime(df["Order Date"])
df["Ship Date"] = pd.to_datetime(df["Ship Date"])
```

```python
df.info()
```

```python
df["Order Year"] = df["Order Date"].dt.year
```

```python
df["Order Month"] = df["Order Date"].dt.month
```

```python
df["Month Name"] = df["Order Date"].dt.month_name()
```

```python
df["Quarter"] = df["Order Date"].dt.quarter
```

```python
df["Shipping Days"] = (
    df["Ship Date"] - df["Order Date"]
).dt.days
```

```python
df["Profit Margin %"] = (
    df["Profit"] / df["Sales"]
) * 100
```

```python
print("Total Sales:", df["Sales"].sum())

print("Total Profit:", df["Profit"].sum())

print("Total Orders:", df["Order ID"].nunique())

print("Total Customers:", df["Customer ID"].nunique())
```

```python
profit_margin = (
    df["Profit"].sum()
    /
    df["Sales"].sum()
) * 100

print(profit_margin)
```

```python
country_sales = (
    df.groupby("Country")["Sales"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

country_sales
```

```python
product_sales=df.groupby("Product Name")["Sales"].sum().sort_values(ascending=False).head(10)
product_sales
```

```python
df.groupby("Category").agg({
    "Sales":sum,
    "Profit":"sum"
}).sort_values(
    by="Sales",ascending=False
)
```

```python
df.groupby("Segment").agg({
    "Sales":"sum",
    "Profit":"sum"
})
```

```python
df.groupby("Region").agg({
    "Sales":"sum",
    "Profit":"sum"
}).sort_values(
    by="Sales",
    ascending=False
)
```

```python
df.groupby("Discount").agg({
    "Sales":"sum",
    "Profit":"sum"
})
```

```python
top_customers=(
    df.groupby("Customer Name").agg({
        "Sales":"sum",
        "Profit":"sum"
    }
    ).sort_values(
        by="Sales",
        ascending=False
    ).head(10)
)
top_customers
```

```python
top_products = (
    df.groupby("Product Name")
      .agg({
          "Sales":"sum",
          "Profit":"sum"
      })
      .sort_values(
          by="Sales",
          ascending=False
      )
      .head(10)
)

top_products
```

```python
df["Discount Band"] = pd.cut(
    df["Discount"],
    bins=[-1,0,0.1,0.2,0.3,1],
    labels=[
        "No Discount",
        "0-10%",
        "10-20%",
        "20-30%",
        "30%+"
    ]
)
```

```python
discount_analysis = (
    df.groupby("Discount Band")
      .agg({
          "Sales":"sum",
          "Profit":"sum"
      })
)

discount_analysis
```
